```python
from google.colab import drive
drive.mount('/content/drive')

import bz2

# Path to the file
file_path = '/content/drive/MyDrive/Colab Notebooks/test.ft.txt.bz2'

# Unzip the .bz2 file and read the first few lines
with bz2.open(file_path, 'rt') as bz_file:
    lines = [bz_file.readline() for _ in range(5)]

# Show the first few lines to check the content
for line in lines:
    print(line)

# Reading the unzipped file completely
reviews = []
with bz2.open(file_path, 'rt') as bz_file:
    for line in bz_file:
        reviews.append(line.strip())

# Show the first few reviews to inspect
for i in range(5):
    print(reviews[i])

# Separate labels and review texts
data = []
for review in reviews:
    parts = review.split(" ", 1)  # Split on the first space
    label = parts[0]              # First part is the label
    text = parts[1] if len(parts) > 1 else ''  # Second part is the review text
    data.append((label, text))

# Display first 5 samples (label and text)
for i in range(5):
    print(data[i])

# Install Google Gemini API
!pip install -q -U google-generativeai

# Necessary packages
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

# Used to securely store your API key
from google.colab import userdata
import google.generativeai as genai
import os

# Configure Google Gemini with your API key
# Or use `os.getenv('GOOGLE_API_KEY')` to fetch an environment variable.
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GOOGLE_API_KEY)


model = genai.GenerativeModel('gemini-pro')

# Define function to get sentiment using Gemini
def get_sentiment_gemini(review_text):
    try:
        # Perform sentiment analysis using Gemini
        response = model.generate_content(f"Perform sentiment analysis for the following review: {review_text}")
        # Check if the response contains valid text
        if response and hasattr(response, 'text'):
            return response.text
        else:
            return "No valid response received from Gemini."
    except Exception as e:
        # Handle potential API or connection errors
        return f"An error occurred: {str(e)}"

# Perform sentiment analysis on the first 5 reviews
import time

for label, review_text in data[:5]:
    sentiment_gemini = get_sentiment_gemini(review_text)
    print(f"Review: {review_text}")
    print(f"Label: {label}")
    print(f"Gemini Sentiment Analysis: {sentiment_gemini}")
    print("-" * 40)

    time.sleep(2)  # Delay to avoid rate limits
```
